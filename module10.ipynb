# Dogs vs Cats

This notebook uses `DogsCats` class to create dataset, buid model,train model and test the model. These are the steps.
- Step 1: Create dataset folders
- Step 2: Create dataset objects
- Step 3: Build the model
- Step 4: Display the model summary
- Step 5: Train the model
- Step 6: Predict test data

### Create Dataset

```python
from dogs_cats import DogsCats

# Initialize DogsCats object
dc = DogsCats()

# Step 1: Create dataset folders
dc.make_dataset_folders('valid', 0, 2400)
dc.make_dataset_folders('train', 2400, 12000)
dc.make_dataset_folders('test', 12000, 12500)

# Step 2: Create dataset objects
dc.make_dataset()

```

```python
# Step 3: Build the model
dc.build_network(augmentation=True)
```

```python
# Step 4: Display the model summary
dc.model.summary()
```

```python
# Step 5: Train the model
dc.train('model.shimil-shijo.dogcats.keras')
```

```python
# Step 6: Predict
dc.load_model('model.shimil-shijo.dogcats.keras')
dc.predict('dogs-vs-cats\\test\cat\cat.12003.jpg')
```
